```python
#‚Ä¢	PASO 3 ACTIVIAD 4 Evaluaci√≥n de factores que influyen en variables clave
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Copia del dataframe
df = df_cleaned.copy()

# Configuraci√≥n de estilo para las gr√°ficas
sns.set_style("whitegrid")

# Gr√°fico de viol√≠n: Distribuci√≥n de ansiedad y depresi√≥n seg√∫n Instrumentalist y Composer
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.violinplot(
    x="Instrumentalist",
    y="Anxiety",
    hue="Instrumentalist",
    data=df,
    ax=axes[0],
    inner="quartile",
    palette="muted",
    dodge=False,
    legend=False
)
axes[0].set_title("Distribuci√≥n de Ansiedad seg√∫n Instrumentalist")
axes[0].set_xlabel("Toca un instrumento (0 = No, 1 = S√≠)")
axes[0].set_ylabel("Nivel de Ansiedad")

sns.violinplot(
    x="Composer",
    y="Depression",
    hue="Composer",
    data=df,
    ax=axes[1],
    inner="quartile",
    palette="muted",
    dodge=False,
    legend=False
)
axes[1].set_title("Distribuci√≥n de Depresi√≥n seg√∫n Composer")
axes[1].set_xlabel("Compone m√∫sica (0 = No, 1 = S√≠)")
axes[1].set_ylabel("Nivel de Depresi√≥n")

plt.tight_layout()
plt.show()

# Correlaciones entre salud mental e Instrumentalist/Composer
correlations = df[["Instrumentalist", "Composer", "Anxiety", "Depression"]].corr()
print("\n\nCorrelaciones entre salud mental e Instrumentalist/Composer:")
print(correlations)

# An√°lisis en 3 p√°rrafos de la gr√°fica de violines

analysis_text = """
En el primer viol√≠n (gr√°fico de la izquierda), se observa la distribuci√≥n de los niveles de ansiedad
seg√∫n si el individuo toca o no un instrumento. Las partes m√°s anchas del viol√≠n representan una mayor
concentraci√≥n de participantes con cierto nivel de ansiedad. La l√≠nea punteada en el centro indica la mediana,
y las l√≠neas m√°s finas muestran los cuartiles, lo que nos permite ver que, para ambas categor√≠as (0 y 1),
los valores de ansiedad se concentran en un rango relativamente similar, aunque existen algunos valores
m√°s altos que se extienden por encima de la mediana.

En el segundo viol√≠n (gr√°fico de la derecha), se ilustra la distribuci√≥n de los niveles de depresi√≥n
en funci√≥n de si la persona compone m√∫sica (Composer). De nuevo, la anchura del viol√≠n refleja la densidad
de participantes, y la l√≠nea punteada central se√±ala la mediana. Aqu√≠ tambi√©n se percibe una distribuci√≥n
similar entre quienes componen (1) y quienes no (0), lo cual sugiere que no hay grandes diferencias en
los niveles de depresi√≥n promedio, aunque se aprecian algunos valores extremos.

En conjunto, ambos violines indican que, en esta muestra de datos, tocar un instrumento o componer m√∫sica
no se asocia claramente con diferencias marcadas en la ansiedad o la depresi√≥n. Aun as√≠, estos resultados
son exploratorios y ser√≠a recomendable complementarlos con pruebas estad√≠sticas (como ANOVA o pruebas
no param√©tricas) para determinar si las diferencias observadas son estad√≠sticamente significativas.
"""

print(analysis_text)

```
